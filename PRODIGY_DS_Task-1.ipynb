```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
population_data = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_5551506.csv', skiprows=4)
```

```python
population1_data = pd.read_csv('Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_5551506.csv', skiprows=4)
```

```python
population2_data = pd.read_csv('Metadata_Indicator_API_SP.POP.TOTL_DS2_en_csv_v2_5551506.csv')
```

```python
# Remove unnecessary columns
population_data = population_data[['Country Name', '2019']]
```

```python
# Rename columns
population_data.columns = ['Country', 'Population']
```

```python
# Drop rows with missing values
population_data = population_data.dropna()
```

```python
# Convert population to numeric
population_data['Population'] = pd.to_numeric(population_data['Population'])
```

```python
# Histogram - Distribution of population
plt.hist(population_data['Population'] / 1e6, bins=20, edgecolor='black')
plt.xlabel('Population (in millions)')
plt.ylabel('Frequency')
plt.title('Distribution of Population')
plt.show()
```

```python
# Bar chart - Top 10 countries by population
top_10_countries = population_data.nlargest(10, 'Population')
plt.bar(top_10_countries['Country'], top_10_countries['Population'] / 1e6)
plt.xlabel('Country')
plt.ylabel('Population (in millions)')
plt.title('Top 10 Countries by Population')
plt.xticks(rotation=45)
plt.show()
```
